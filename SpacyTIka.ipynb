```
from tika import parser
import spacy
nlp = spacy.load('en_core_web_sm')
```

```
parsedPDF = parser.from_file('/content/drive/MyDrive/Colab Notebooks/Spacy/RDouglas.pdf')
```

```
parsedPDF
```

```
parsedPDF['metadata']
```

```
pdf_content=parsedPDF['content']
```

```
pdf_content
```

```
doc = nlp(pdf_content)
```

```
doc[1:4]
```

## Tokenization 

```
for token in doc:
  print(token.text)
```

```
#Print the tokens which are like numbers
for token in doc:
  if token.like_num:
    print(token)
```

```
#detecting email addresses
for token in doc:
  if token.like_email:
    print(token.text)
```

```
#detecting url
for token in doc:
  if token.like_url:
    print(token.text)
```

```
#detecting upper case
for token in doc:
  if token.is_upper:
    print(token.text)
```

```
for token in doc:
  print(token.i,"|,", token.text,"|", token.is_alpha, "|", token.is_upper)
```

```
for token in doc:
  if token.is_upper == True:
    print(token.i, token.text)
```

```
for token in doc:
  if token.__len__:
    print(token.i, token.text)
```
